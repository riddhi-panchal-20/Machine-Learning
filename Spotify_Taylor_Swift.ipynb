```python
name = "Taylor Swift"
result = sp.search(name)
result['tracks']['items'][0]['artists']
```

```python
artist_uri = result['tracks']['items'][0]['artists'][0]['uri']
sp_albums = sp.artist_albums(artist_uri, album_type='album')
album_names = []
album_uris = []
for i in range(len(sp_albums['items'])):
    album_names.append(sp_albums['items'][i]['name'])
    album_uris.append(sp_albums['items'][i]['uri'])

print(album_names)
print(album_uris)
```

```python
song_info = []  # To store track information

# Iterate through the album URIs
for album_uri in album_uris:
    # Fetch the tracks for the album
    sp_tracks = sp.album_tracks(album_uri)

    # Iterate through the tracks and extract the desired information
    for track in sp_tracks['items']:
        track_info = {
            'track_id': track['id'],
            'track_name': track['name'],
            'track_uri': track['uri']
        }
        song_info.append(track_info)

# Print the track information
for track in song_info:
    print("Track ID:", track['track_id'])
    print("Track Name:", track['track_name'])
    print("Track URI:", track['track_uri'])

```
